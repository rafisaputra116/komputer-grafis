```python
import numpy as np
import matplotlib.pyplot as plt

plt.axis([0, 150, 100, 0])
r = 40
alpha1 = np.radians(0)
alpha2 = np.radians(360)
dalpha = np.radians(2)
xc = 75
yc = 50
plt.scatter(xc, yc, s=10, color='k')

for alpha in np.arange(alpha1, alpha2, dalpha):
    x = xc + r * np.cos(alpha)
    y = yc + r * np.sin(alpha)
    plt.scatter(x, y, s=5, color='k')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.axis([-10, 140, 90, -10])

# Garis merah dengan ketebalan 2 (solid default)
plt.plot([40, 100], [20, 20], linewidth=2, color='r')

# Garis hijau dengan ketebalan 4 dan gaya titik-titik
plt.plot([40, 100], [30, 30], linewidth=4, color='g', linestyle=':')

# Garis biru dengan ketebalan 6 dan gaya garis putus-putus
plt.plot([40, 100], [40, 40], linewidth=6, color='b', linestyle='--')

# Garis hitam dengan ketebalan 2 dan gaya garis titik-strip
plt.plot([40, 100], [50, 50], linewidth=2, color='k', linestyle='-.')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.axis([-10, 140, 90, -10])

# Panah merah dengan kepala panah kecil
plt.arrow(40, 20, 60, 0, linewidth=1, color='r',
          head_length=5, head_width=3)

# Panah hijau dengan gaya titik-titik dan kepala panah besar
plt.arrow(40, 30, 60, 0, linewidth=1, color='g', linestyle=':',
          head_length=10, head_width=5)

# Panah biru dengan gaya garis putus-putus dan kepala panah sedang
plt.arrow(40, 40, 60, 0, linewidth=1, color='b', linestyle='--',
          head_length=8, head_width=4)

# Panah hitam tebal dengan garis solid dan kepala panah sedang
plt.arrow(40, 50, 60, 0, linewidth=4, color='k', linestyle='-',
          head_length=8, head_width=3)

plt.show()
```
